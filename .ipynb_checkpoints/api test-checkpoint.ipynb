# DART 에서 종목코드-회사코드 쌍 가져오기
* 변수
    * info = {stockCode: corpCode, corpName}

```python
import requests
from zipfile import ZipFile
from io import BytesIO
import xml.etree.ElementTree as et
import pandas as pd
```

```python
url = 'https://opendart.fss.or.kr/api/corpCode.xml'
key = '6e2b668b16e3457e032a8c7c03f5ee3f7af9d221'
params = {'crtfc_key': key}
res = requests.get(url, params)
```

```python
zo = ZipFile(BytesIO(res.content))

xmlfile_objs = {name: zo.read(name) for name in zo.namelist()}
xml_str = xmlfile_objs['CORPCODE.xml'].decode('utf-8')
```

```python
# XML String을 가져와서 DataFrame에 담기
xroot = et.fromstring(xml_str)

df_cols = ["corp_code", "corp_name", "stock_code", "modify_date"]
rows = []

for node in xroot: 
    res = []
    for el in df_cols[0:]: 
        if node is not None and node.find(el) is not None:
            res.append(node.find(el).text)
        else: 
            res.append(None)
    rows.append({df_cols[i]: res[i] 
                  for i, _ in enumerate(df_cols)})

df = pd.DataFrame(rows, columns=df_cols)
```

```python
stock = df[df['stock_code'] != ' ']
stock
stock.to_csv('codes.csv')
```

## naver finalce API (REST) 로 주가 데이터 받아오는 함수

```python
from urllib import parse
from ast import literal_eval
import requests

def get_sise(code, start_time, end_time, time_from='day') :
    get_param = {
    	'symbol':code,
	    'requestType':1,
	    'startTime':start_time,
	    'endTime':end_time,
	    'timeframe':time_from
    }
    get_param = parse.urlencode(get_param)
    url="https://api.finance.naver.com/siseJson.naver?%s"%(get_param)
    response = requests.get(url)
    return literal_eval(response.text.strip())
```

```python
def GetDatas(stock_codes):
    columns = ['name', 'corp_code', 'sotck_code', 'ROE', 'ROI', 'PBR', 'PER', 'GP/A', 'price', 'cap', 'momentum']
    df = pd.DataFrame(columns=columns)
    for stock_code in stock_codes:
        infos = GetStockInfo(stock_code)
        print(infos)
        df = df.append({'name': 'dd'}, ignore_index=True)
    
    return df
```

```python
def GetStockInfo(stock_code):
    '''
        get stock infos by corp code
        
        corp_code
        재무제표 - 손익계산서, 재무상태표, 현금흐름표
        과거 가격 (for momentum)
        
    '''
    
    corp_code = GetCorpCodeByStockCode(stock_code)
    corp_name = GetCorpNameByStockCode(stock_code)
    # 재무제표 데이터 가져오기
    report = GetReport(corp_code)
    print(corp_name, report)

    # 가격 데이터 가져오기
    prices = GetPrices(stock_code, '20210101', '20220505', 'month')
    # print(prices)
    opens = [price[1] for price in prices]
    
    return report, opens
```

```python
def GetCorpCodeByStockCode(stock_code):
    corp_code = stock[stock['stock_code']==stock_code]['corp_code']
    return corp_code

def GetStockCodeByCorpCode(corp_code):
    stock_code = stock[stock['corp_code']==corp_code]['stock_code']
    return stock_code

def GetCorpNameByStockCode(stock_code):
    corp_name = stock[stock['stock_code']==stock_code]['corp_name']
    return corp_name

def GetReport(corp_code):
    url = 'https://opendart.fss.or.kr/api/fnlttSinglAcntAll.json'
    key = '6e2b668b16e3457e032a8c7c03f5ee3f7af9d221'
    params = {'crtfc_key': key, 'corp_code': corp_code, 'bsns_year': '2021', 'reprt_code': '11013', 'fs_div': 'CFS'}
    res = requests.get(url, params)
    j = res.json()
    data = j['list']
    data = ConvertReport(data)
    return data


def GetCaps(corp_codes):
    '''
        시가 총액 구하는 함수.
        
        params:
            corp_codes: list (8 digit corp code)
            
        return:
            dict: {corp_code: cap}
    '''
    
    url = 'https://opendart.fss.or.kr/api/stockTotqySttus.json'
    key = '6e2b668b16e3457e032a8c7c03f5ee3f7af9d221'
    params = {'crtfc_key': key, 'corp_code': corp_codes, 'bsns_year': '2022', 'reprt_code': '11013'}
    res = requests.get(url, params)
    result = res.json()['list']
    
    d = {}
    for r in result:
        if r['se'] == '합계':
            print(r['distb_stock_co'])
            code = r['corp_code']
            d[code] = r['distb_stock_co']
    return d


def GetPrices(stock_code, start_time, end_time, time_from='day') :
    get_param = {
        'symbol':stock_code,
        'requestType':1,
        'startTime':start_time,
        'endTime':end_time,
        'timeframe':time_from
    }
    get_param = parse.urlencode(get_param)
    url="https://api.finance.naver.com/siseJson.naver?%s"%(get_param)
    response = requests.get(url)
    return literal_eval(response.text.strip())


def ConvertReport(report):
    content = {}
    columns = ['당기순이익(손실)', '자본총계', '자산총계', '유동자산', '비유동자산', '매출총이익']
    for rep in report:
        if rep['account_nm'] in columns:
            # print(rep['account_nm'], rep['frmtrm_amount'])
            content[rep['account_nm']] = rep['frmtrm_amount']
            
    return content
```
